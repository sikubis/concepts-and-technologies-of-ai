```python
import matplotlib.pyplot as plt

# Daily time (in hours): [study, entertainment, sleep]
time_data = [
    (3.5, 2.0, 7.0), (5.0, 1.5, 6.5), (2.5, 3.0, 8.0),
    (4.0, 2.0, 6.0), (1.5, 4.5, 9.0), (3.0, 2.5, 7.5),
    (5.5, 1.0, 6.0), (2.0, 3.5, 8.5), (4.5, 2.0, 7.0),
    (3.0, 3.0, 7.5), (6.0, 1.5, 6.0), (2.5, 4.0, 8.0),
    (4.0, 2.5, 7.0), (5.0, 2.0, 6.5), (3.5, 2.5, 7.0)
]

#task 1
low_study_time = []
moderate_study_time = []
high_study_time = []

for study, entertainment, sleep in time_data:
    if study < 3:
        low_study_time.append(study)
    elif 3 <= study <= 5:
        moderate_study_time.append(study)
    else:
        high_study_time.append(study)

print("Low study:", low_study_time)
print("Moderate study:", moderate_study_time)
print("High study:", high_study_time)

#task 2
print("-" * 50)
print("Days with low study time:", len(low_study_time))
print("Days with moderate study time:", len(moderate_study_time))
print("Days with high study time:", len(high_study_time))

#task 3
study_minutes = []
multiplier = 60

for day in time_data:
    study_hours = day[0]
    minutes = study_hours * multiplier
    study_minutes.append(minutes)

print("-" * 50)
print("Study time in minutes:", study_minutes)

#task 4
study_hours = []
entertainment_hours = []
sleep_hours = []

for study, entertainment, sleep in time_data:
    study_hours.append(study)
    entertainment_hours.append(entertainment)
    sleep_hours.append(sleep)

num_days = len(time_data)
avg_study_hours = sum(study_hours) / num_days
avg_entertainment_hours = sum(entertainment_hours) / num_days
avg_sleep_hours = sum(sleep_hours) / num_days

print("-" * 50)
print(f"Average hours spent studying: {avg_study_hours:.2f} hours")
print(f"Average hours spent on entertainment: {avg_entertainment_hours:.2f} hours")
print(f"Average hours spent sleeping: {avg_sleep_hours:.2f} hours")

#task 5
print("-" * 50)
study_hours = [3.5, 5.0, 2.5, 4.0, 1.5, 3.0, 5.5, 2.0, 4.5, 3.0, 6.0, 2.5, 4.0, 5.0, 3.5]
sleep_hours = [7.0, 6.5, 8.0, 6.0, 9.0, 7.5, 6.0, 8.5, 7.0, 7.5, 6.0, 8.0, 7.0, 6.5, 7.0]

plt.figure(figsize=(8, 6))
plt.scatter(sleep_hours, study_hours, alpha=0.5)
plt.title('Study Hours vs. Sleep Hours Pattern', fontsize=14)
plt.xlabel('Study Hours', fontsize=12)
plt.ylabel('Sleep Hours', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
```

```python
#Recursion:
#task 1
def sum_nested_list(nested_list):
  total = 0
  for i in nested_list:
     if isinstance(i, list):
      total += sum_nested_list(i)
     else:
      total += i
  return total

data = [1, [2, [3, 4], 5], 6, [7, 8]]
print(sum_nested_list(data))

#task 2
print("-" * 50)
def generate_permutations(s):
  results = []
  if len(s) == 1:
    return [s]

  else:
    for index in range(len(s)):
      current_char = s[index]
      remaining_chars = s[:index] + s[index + 1:]
      for permutation in generate_permutations(remaining_chars):
        new_string = current_char + permutation
        if new_string not in results:
                results.append(new_string)

  return results

print(generate_permutations("abc"))
print(generate_permutations("aab"))

#task3
print("-" * 50)
def calculate_directory_size(directory):
    total_size = 0

    for name, value in directory.items():
        if isinstance(value, int):
            total_size += value

        else:
            total_size += calculate_directory_size(value) #going inside the subdirectory and calling the function agaun

    return total_size

directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,

    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },

    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

print(calculate_directory_size(directory_structure))
```

```python
#Dynamic Programming:
#task1
def min_coins(coins, amount):
  dp = [float('inf')] * (amount + 1)
  dp[0] = 0

  for coin in coins:
    for i in range(coin, amount + 1):
      dp[i] = min(dp[i], dp[i - coin] + 1)
  return dp[amount] if dp[amount] != float('inf') else -1

print(min_coins([1, 2, 5], 11))

#task2
def longest_common_subsequence(s1, s2):
    m, n = len(s1), len(s2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
    return dp[m][n]

s1 = "abcde"
s2 = "ace"

print(longest_common_subsequence(s1, s2))

#task3
def knapsack(weights, values, capacity):
    dp = [0] * (capacity + 1)

    for i in range(0, len(weights)):
        for j in range(weights[i], capacity + 1):
            dp[j] = max(values[i] + dp[j - weights[i]], dp[j])
    return dp[capacity]

weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7

print(knapsack(weights, values, capacity))
```
