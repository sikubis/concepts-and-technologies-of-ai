```python
import pandas as pd

# 3.2 Exploratory Data Analysis Practice Exercise - 1.
df = pd.read_csv("titanic.csv") # loading csv
df.info()   # checking dataframe info using info.()

df["Age"] = df["Age"].fillna(df["Age"].median()) # fillna() replaces all the empty cells with a given value. in this case, replaces with median value
df["fare_per_year"] = df["Fare"] / df["Age"]   # new column containing the fare divided by the age of the passenger

high_fare_age = df[df["fare_per_year"] > 5] # subsetting rows where fare_per_year is higher than 5
high_fare_age_srt = high_fare_age.sort_values(by="fare_per_year", ascending=False) # sorting by fare_per_year with ascending=False which means descending
result = high_fare_age_srt[["Name", "fare_per_year"]] # selecting only name and fare_per_year columns and saving them to result
print(result) # Allison, Master. Hudson Trevor had the highest fare relative to age, with a fare per year of approximately 164.73


df["fare_per_class"] = df["Fare"] / df["Pclass"]   # new column containing the fare divided by the class of the passenger

adult_males = df[(df["Sex"] == "male") & (df["Age"] >= 18)] # filtering out passengers that are male and above 18
adult_males_srt = adult_males.sort_values(by="fare_per_class", ascending=False) # sorting by fare_per_class with ascending=False which means descending
result = adult_males_srt[["Name", "Age", "fare_per_class"]] # selecting only name, age and fare_per_class columns and saving them to result
print(result) # Lesurer, Mr. Gustave J and Cardeza, Mr. Thomas Drake Martinez paid the highest fare relative to their class, with a fare per class of approximately 512.33


```

```python
import pandas as pd

# 3.3 Exploratory Data Analysis with Group-by Method Practice Exercise:
df = pd.read_csv("titanic.csv") # loading csv
df.info()   # checking dataframe info using info.()

total_fare = df['Fare'].sum()    # total Fare paid across all passengers in the Titanic dataset
first_class = df[df["Pclass"] == 1]   # filtering out first class passengers
total_fare_first_class = first_class["Fare"].sum()     # summing total fare paid of those in first class
second_class = df[df["Pclass"] == 2]   # filtering out second class passengers
total_fare_second_class = second_class["Fare"].sum()  # summing total fare paid of those in second class
third_class = df[df["Pclass"] == 3]   # filtering out third class passengers
total_fare_third_class = third_class["Fare"].sum()  # summing total fare paid of those in third class
fare_totals_by_class = [
    total_fare_first_class,
    total_fare_second_class,
    total_fare_third_class
] # sum of fares sorted by pclass
fare_proportions = [
    total_fare_first_class / total_fare,
    total_fare_second_class / total_fare,
    total_fare_third_class / total_fare
] # revenue proportion by class
print(fare_proportions)
# first class passengers contributed about 63.3% of the total fare
# second class passengers contributed about 13.2% of the total fare
# third class passengers contributed about 23.4% of the total fare

df["Age"] = df["Age"].fillna(df["Age"].median()) # fillna() replaces all the empty cells with a given value. in this case, replaces with median value

df.loc[df["Age"] < 18, "age_group"] = "child" # .loc is a pandas method that is used to access or modify rows and cols through labels or bool conditions
df.loc[(df["Age"] >= 18) & (df["Age"] < 65), "age_group"] = "adult"
df.loc[df["Age"] >= 65, "age_group"] = "senior"
total_passengers = len(df) # calculating total no. of passengers
age_group_counts = df["age_group"].value_counts() # .value_counts() counts how many times does a unique value repeats
age_group_proportions = age_group_counts / total_passengers # the proportion of passengers in each age group
age_group_percent = age_group_proportions * 100
print(age_group_percent) # approximately 86.10% were adults, 12.69% of passengers were children and 1.23% were seniors
```
